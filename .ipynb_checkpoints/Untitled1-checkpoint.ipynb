```python
import osmnx as ox
import geopandas as gpd
```

```python
study_area_filename = 'shapes/study_area/study_area_buffer.shp'
study_area = gpd.read_file(study_area_filename).to_crs('EPSG:4326')
study_area_polygon = study_area.iloc[0]['geometry']
nodes, edges = ox.graph_to_gdfs(ox.graph_from_polygon(study_area_polygon, network_type='walk', simplify=True))
```

```python
import matplotlib.pyplot as plt

def plot_graph(gdf):
    fig, ax = plt.subplots(figsize=(14, 14))
    gdf.plot(ax=ax)
    
plot_graph(edges)
```

```python
edges['osmid'] = edges[['u', 'v']].apply(lambda x: str(x[0]) + '-' + str(x[1]), axis=1)
edges_no_duplicates = edges.sort_values(['osmid','length']).drop_duplicates('osmid', keep='first')
```
