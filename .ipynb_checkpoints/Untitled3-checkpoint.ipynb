```python
import osmnx as ox
import geopandas as gpd
import networkx as nx
```

```python
study_area_filename = 'shapes/study_area/study_area.shp'
study_area = gpd.read_file(study_area_filename).to_crs('EPSG:4326')
study_area_polygon = study_area.iloc[0]['geometry']
    
```

```python
bike_network = ox.graph_from_polygon(study_area_polygon, network_type='bike', simplify=False)
```

```python
bike_nodes, bike_edges = ox.graph_to_gdfs(bike_network)
```

```python
bike_nodes.to_file('shapes/bike/nodes.shp')
bike_edges.to_file('shapes/bike/edgdes.shp')
```

```python
infra = 'way["highway"~"motorway|primary"]'
walk_network = ox.graph_from_polygon(study_area_polygon, network_type='walk',
                                     infrastructure='way["sidewalk"]',
                                     simplify=False)
```

```python
walk_nodes, walk_edges = ox.graph_to_gdfs(walk_network)
```

```python
def reverse_osmids(gdf):
    two_ways_edges = gdf[~gdf['oneway']].copy()
    for i, u, v, osmid in two_way_edges.itertuples():
        if not osmid + '-R' in gdf.osmid:
            gdf.loc[gdf.u == v & gdf.v == u, 'osmid'] = osmid + '-R'
    return gdf
        
        
```
