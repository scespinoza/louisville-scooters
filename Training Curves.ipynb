```python
import pickle
```

```python
with open('weights/monday_3.0_test.pickle', 'rb') as file:
    model = pickle.load(file)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
plt.plot(model.history['rewards'], label='batch loss')
n = 10

moving_loss = np.convolve(model.history['rewards'], np.ones((n,))/n, mode='valid', )
plt.plot(moving_loss, label='moving loss')
plt.legend()
```
