```python
import pickle
%config InlineBackend.figure_format = 'retina'
```

```python
with open('weights/monday_3.0_gpu.pickle', 'rb') as file:
    model = pickle.load(file)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')

n = 10
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(model.history['rewards'][10:], label='episode rewards', color='green')
moving_loss = np.convolve(model.history['rewards'][10:], np.ones((n,))/n, mode='valid')
ax.plot(moving_loss, label='moving rewards', color='orange')
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
n = 24
ax.plot(model.history['batch_loss'], label='batch loss')
moving_loss = np.convolve(model.history['batch_loss'], np.ones((n,))/n, mode='valid')
ax.plot(moving_loss, label='moving loss', color='orange')
plt.legend()
```
